```python
store = pd.DataFrame([[100,70,30],
                     [100,90,10],
                     [200,100,100],
                     [150,80,70]] ,
                   columns = ['입고량','출고량','재고량'],
                   index = ['연필','노트','지우개','볼펜'])
store
```

```python
plt.figure(figsize = (8, 4))
plt.plot(store.index, store['입고량'])
plt.plot(store.index, store['출고량'])
plt.plot(store.index, store['재고량'])
plt.grid(color = 'red', linestyle = ':', alpha = 0.5)
plt.yticks([50, 100, 150, 200, 250])
plt.show()
```

```python
store.index.name = "상품명"
store
```
