```python
df = sns.load_dataset('iris')
df
```

```python
tips['won'] = tips['tip'].apply(lambda x : x * 1150)
tips
```

```python
def gender(x):
    if x == 'Male':
        x = 'man'
    elif x == 'Female':
        x = 'woman'
    return x

tips['sex'] = tips['sex'].apply(gender)
tips
```

```python
tips['sex'] = tips['sex'].apply(lambda x : x[0].upper() + x[1:])
tips
```

```python
def addBill(x):
    x['total_bill'] = '$' + str(x['total_bill'])
    x['tip'] = '$' + str(x['tip'])
    return x
 
tips = tips.apply(addBill, axis = 1)
tips
```
