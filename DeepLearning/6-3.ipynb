```python
import requests
from bs4 import BeautifulSoup

for i in range(1, 5):
    url = f'http://www.yes24.com/24/category/bestseller?CategoryNumber=00{i}'
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    div = soup.find('div', {"id" : "bestList"})
    ol = div.find('ol')
    for j in range(1, 4):
        li = ol.find('li', {"class" : f"num{j}"})
        src = li.img['src']
        img_src = requests.get(src)
        img_src.raise_for_status()        
        with open('cover{}-{}.jpg'.format(i, j), 'wb') as f:
            f.write(img_src.content)

print('-----프로그램완료-----')
```
