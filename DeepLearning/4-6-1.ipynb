```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.naver.com/'
res = requests.get(url)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'lxml')

divs = soup.find_all('div', attrs = {'class' : 'cjs_channel_title _category_title_a'})
for i, div in enumerate(divs):
    print(i + 1, div.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.naver.com/'
res = requests.get(url)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'lxml')

uls = soup.find('ul', attrs = {'class' : 'cjs_channel_list _category_list_wrapper_a'})
divs = uls.find_all('div', attrs = {'class' : 'cjs_channel_name'})

for i, div in enumerate(divs):
    print(i + 1, div.text)
```
